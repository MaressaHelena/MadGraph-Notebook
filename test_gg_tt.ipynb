```python
df = read_lhe_uproot('test_gg_tt/Events/run_01/unweighted_events.root')
df
```

```python
df['cos_theta'] = df.Pz / np.sqrt(df.Pz ** 2 + df.PT ** 2)
selection = ((df.Status == 1) & (df.PID == 6))
df_sel = df [selection]
df_sel
```

```python
n_events = len(df_sel)
weights = np.ones(n_events)*(1/n_events)
weights
```

```python
fig = plt.figure (figsize=(12,10))
plt.hist(df_sel['cos_theta'], histtype='step', bins=100)
plt.xlabel(r'$\cos\theta$')
plt.ylabel('Número de Eventos')
```
